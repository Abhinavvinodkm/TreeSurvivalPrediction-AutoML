```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 200)
import pickle
```

```python
df = pd.read_csv('Tree_Data.csv')
```

```python
df.head(5)
```

# Columns selection

```python
columns_to_keep = ['Plot', 'Subplot', 'Species', 'Light_Cat', 'Core',
       'Soil', 'Adult', 'Sterile', 'Conspecific', 'Myco', 'SoilMyco',
       'AMF', 'Phenolics', 'Lignin', 'NSC', 'Alive']

df = df[columns_to_keep].copy()
```

# Preprocessing

```python
df_encoded = df.copy()
```

```python
# One hot encoding

df_encoded  = pd.get_dummies(df_encoded, columns=['Subplot', 'Species', 'Soil', 'Adult', 'Conspecific', 'SoilMyco'])

df_encoded.head(5)
```

```python
# Mapping categorical data

map_dict = {"Low": 0, "Med": 1, "High": 2}
df_encoded['Light_Cat'] = df_encoded['Light_Cat'].map(map_dict)

map_dict = {2017:1, 2016:0}
df_encoded['Core'] = df_encoded['Core'].map(map_dict)

map_dict = {'Non-Sterile':0, 'Sterile':1}
df_encoded['Sterile'] = df_encoded['Sterile'].map(map_dict)
df_encoded.head(5)

map_dict = {'AMF': 1, 'EMF' : 0}
df_encoded['Myco'] = df_encoded['Myco'].map(map_dict)

df_encoded.head(5)
```

```python
# dealing with NaNs and mapping values in the 'Alive' column

df_encoded['Alive'].fillna(0, inplace = True)
df_encoded['Alive'] = df_encoded['Alive'].replace({'X': 1})
df_encoded.head(10)
```

```python
df_encoded.to_csv('Tree_Data_Preprocessed.csv', index=False)
```
